```python
import pandas as pd

import numpy as np
from scipy import interpolate
df = pd.read_csv("Birth.csv",engine = 'python')
df.head()

df2 = df.drop(columns = ['Country Name'])
df2= df2.interpolate()
df2
```

```python
def prediction1(x,y):
    from sklearn import linear_model   
    
    y = np.array(y)
    x = x.reshape(-1,1)
   
    reg = reg = linear_model.BayesianRidge()
    LM  = reg.fit(x,y) #Note I am passing in x and y in column shape
    p = []
    for i in range (2018,2031):
        p.append(i)
    predict_me = np.array([p])

    predict_me = predict_me.reshape(-1,1)

    score = LM.score(x,y)


    predicted_values = LM.predict(predict_me)
    return predicted_values
```

```python
temp = df2.values.tolist()
x=[]
for i in range(1960,2017):
    x.append(i)
x= np.array(x)  
prediction =[]
z= []
for i in temp:
    z.append(prediction1(x,i))
q = pd.DataFrame(list(map(np.ravel, z))) 
qw = []
for i in range (2018,2031):
        qw.append(i)
q.columns = qw
q.insert(0, 'Country Name', df.ix[:,0])
q.to_csv('Birth predicted.csv')
```
