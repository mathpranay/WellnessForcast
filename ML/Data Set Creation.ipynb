```python
import pandas as pd
import numpy as np
```

```python
b =["Factors"]
for i in range(2018,2031):
    b = np.append(b,i)
b
```

```python
final=pd.DataFrame(columns=b)
final.set_index('Factors')
#print(final)

factors=["AccessToElectricity","CrudeBirthRate","FemalePopulationRatio","LifeExpectancy","DeathRate","CO2perCapita","InfantMortalityRate","Food"]
```

```python
q = pd.read_csv("Wellness/Human development index (HDI).csv")
z = q['Country'].values
for j in z:
    for i in range(0,8):
        df = pd.read_csv("Predicted/"+factors[i]+"Predicted.csv",encoding = 'latin-1')
        #print(df)
       # df.set_index('Country Name')
        a = df[df['Country Name']==j.strip()].values        
        if(a.shape==(0,14)):
            b=df.mean().values
            final.loc[i]=np.append([''],b)
            final.iloc[i,0]=factors[i]
        else:
            final.loc[i]=a[0]
            final.iloc[i,0]=factors[i]
 
    f = final.T
    f.to_csv('D/'+j.strip()+'.csv',index=False, header=False)
```
