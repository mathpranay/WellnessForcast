```python
from sklearn import svm
from sklearn import preprocessing
df1 = data.drop(columns = ['HDI'])
print(df1)
df10 =pd.DataFrame(data, columns = ['HDI'])
df11 = df1.values
df10= df10.values
print(df11)
#min_max_scaler = preprocessing.MinMaxScaler()
#df11 = min_max_scaler.fit_transform(df11)
print(df11)
#lab_enc = preprocessing.LabelEncoder()
#df11 = lab_enc.fit_transform(df10)
clf = linear_model.LinearRegression()
clf.fit(df11,df10)

```

```python
dtf=pd.read_csv('Human development index (HDI).csv')



b =["Country Name"]
for i in range(2018,2031):
    b = np.append(b,i)

final=pd.DataFrame(columns=b)

for x in dtf['Country'].values:
    dd = pd.read_csv('D/'+x.strip()+'.csv')
    dd = dd.dropna()
    #dd1 =dd.drop(columns = ['HDI'])
    dd2 = dd.values
    #dd2 = min_max_scaler.fit_transform(dd2)
    predict1 = []
    predict1.append(clf.predict(dd2))
    new_row = np.append([x],np.ravel(predict1))
    final.loc[len(final)] = new_row
final
```
