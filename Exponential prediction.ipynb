```python
from scipy import optimize
def prediction1(x,y):    
    params, params_covariance = optimize.curve_fit(func, x, y,p0=[45, 0.6],maxfev = 5000)
    result=[]
    for i in range(59,89):
        result.append([func(i, params[0],params[1])])
        result1=np.array(result)
        predicted_values = result1
    return predicted_values
```
